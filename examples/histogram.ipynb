```python
lena_star = Image("data/lena_star.jpg")
lena_star2 = Image("data/lena_star2.jpg")
lena_box = Image("data/lena_box.jpg")
box = Image("data/box.jpg")
```

```python
print(lena.compare_to(lena, method=ImageDiffMethod.mse))
print(lena.compare_to(lena_star, method=ImageDiffMethod.mse))
print(lena.compare_to(lena_star2, method=ImageDiffMethod.mse))
print(lena.compare_to(lena_box, method=ImageDiffMethod.mse))
print(lena.compare_to(box, method=ImageDiffMethod.mse))
```
