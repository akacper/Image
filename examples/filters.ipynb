```python
from Image import Image
from matplotlib.pyplot import imshow
```

```python
# Filtr tożsamościowy
id_kernel = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])

# Filtr górnoprzepustowy - wyostrzanie obrazu. Filtr wzmacnia szczegóły.
up_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

# Filtr dolnoprzepustowy - rozmycie obrazu. Filtr uśrednia szczegoły.
down_kernel = 1/9 * np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])

# Rozmycie Gaussowskie - wersja 3x3 px
gauss_3_kernel = 1/16 * np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])

# Rozmycie Gaussowskie - wersja 5x5 px
gauss_5_kernel = 1/256 * np.array([[1, 4, 6, 4, 1], [4, 16, 24, 16, 4], [6, 24, 36, 24, 6], [4, 16, 24, 16, 4], [1, 4, 6, 4, 1]])
```

```python
# Operatory Sobela - detekcja krawędzi
sobel0 = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel45 = np.array([[0, 1, 2], [-1, 0, 1], [-2, -1, 0]])
sobel90 = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
sobel135 = np.array([[2, 1, 0], [1, 0, -1], [0, -1, -2]])
```

```python
lena = Image("data/lena.jpg")
gray_lena = lena.to_gray()
```

```python
gray_lena.conv_2d(id_kernel).show_img()
```

```python
f, ax_arr = plt.subplots(1, 5, figsize=(14,10))
ax_arr[0].imshow(lena.conv_2d(id_kernel).data, cmap='gray')
ax_arr[1].imshow(lena.conv_2d(gauss_3_kernel).data, cmap='gray')
ax_arr[2].imshow(lena.conv_2d(gauss_5_kernel).data, cmap='gray')
ax_arr[3].imshow(lena.conv_2d(up_kernel).data, cmap='gray')
ax_arr[4].imshow(lena.conv_2d(down_kernel).data, cmap='gray')
```

```python
f, ax_arr = plt.subplots(1, 5, figsize=(14,10))
ax_arr[0].imshow(gray_lena.conv_2d(id_kernel).data, cmap='gray')
ax_arr[1].imshow(gray_lena.conv_2d(gauss_3_kernel).data, cmap='gray')
ax_arr[2].imshow(gray_lena.conv_2d(gauss_5_kernel).data, cmap='gray')
ax_arr[3].imshow(gray_lena.conv_2d(up_kernel).data, cmap='gray')
ax_arr[4].imshow(gray_lena.conv_2d(down_kernel).data, cmap='gray')
```
