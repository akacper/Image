```python
from Image import BaseImage
from matplotlib.pyplot import imshow
```

# Podstawowe metody

```python
lena = BaseImage("data/lena.jpg")
warstwa_leny = lena.get_layer(0)
warstwa_leny1 = lena.get_layer(1)
warstwa_leny2 = lena.get_layer(2)

import matplotlib.pyplot as plt


f, ax_arr = plt.subplots(1,3)
ax_arr[0].imshow(warstwa_leny.data, cmap='gray')
ax_arr[1].imshow(warstwa_leny1.data, cmap='gray')
ax_arr[2].imshow(warstwa_leny2.data, cmap='gray')
plt.show()
```

# Konwersja
RGB -> HSV

```python
hsv_lena = lena.to_hsv()
print(hsv_lena.color_model)
```

HSV -> RGB

```python
rgb_lena_from_hsv = hsv_lena.to_rgb()
print(rgb_lena_from_hsv.color_model)
rgb_lena_from_hsv.show_img()
```

RGB -> HSI

```python
hsi_lena = lena.to_hsi()
print(hsi_lena.color_model)
```

HSI -> RGB

```python
rgb_lena_from_hsi = hsi_lena.to_rgb()
print(rgb_lena_from_hsi.color_model)
rgb_lena_from_hsi.show_img()
```

RGB -> HSL

```python
hsl_lena = lena.to_hsl()
print(hsl_lena.color_model)
```

HSL -> RGB

```python
rgb_lena_from_hsl = hsl_lena.to_rgb()
print(rgb_lena_from_hsl.color_model)
rgb_lena_from_hsl.show_img()
```
