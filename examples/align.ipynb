```python
from Image import Image, ImageDiffMethod, plt
from matplotlib.pyplot import imshow
```

```python
lena = Image("data/lena.jpg")
```

```python
gray_lena = lena.to_gray()
```

```python
align_gray = gray_lena.align_image(tail_elimination=False)
f, ax_arr = plt.subplots(1, 2, figsize=(14,10))
ax_arr[0].imshow(gray_lena.data, cmap='gray')
ax_arr[1].imshow(align_gray.data, cmap='gray')
```

```python
gray_lena.compare_to(align_gray, method=ImageDiffMethod.mse)
```

```python
aligned = lena.align_image(tail_elimination=False)
aligned_tail = lena.align_image(tail_elimination=True)

f, ax_arr = plt.subplots(1, 3, figsize=(14,10))
ax_arr[0].imshow(lena.data)
ax_arr[1].imshow(aligned.data)
ax_arr[2].imshow(aligned_tail.data)
```
