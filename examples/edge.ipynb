```python
from Image import Image, plt
from matplotlib.pyplot import imshow
import cv2
```

```python
sudoku = Image('data/sudoku.jpg')
```

```python
gray_sudoku = sudoku.to_gray().baseimage_to_image()
task_1 = gray_sudoku.adaptive_threshold().baseimage_to_image()
task_1.show_img()
```

```python
task_2 = task_1.line_detection(rho=0.5, color=(255, 0, 0), thickness=2)
task_2.show_img()
```

```python
lines_image = Image('data/lines.jpg')
gray_lines = lines_image.to_gray()
lines_thresh = gray_lines.lines_thresh().baseimage_to_image()
lines_thresh.show_img()
```

```python
lines_edges = lines_thresh.canny(20, 50, 3).baseimage_to_image()
lines_edges.show_img()
```

```python
lines = lines_edges.hough_lines_p()
len(lines)
```
